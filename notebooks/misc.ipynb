```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
from numpy.polynomial.polynomial import Polynomial

x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 0.5, 2, 1, 1.5])

# Piecewise Cubic Spline (Natural)
spline = CubicSpline(x, y, bc_type='natural')

# Global Cubic Polynomial (fit to all points at once)
coeffs = np.polyfit(x, y, 3)
global_cubic = np.poly1d(coeffs)

xx = np.linspace(0, 4, 200)
plt.plot(x, y, 'o', label='Data')
plt.plot(xx, spline(xx), label='Piecewise Cubic Spline')
plt.plot(xx, global_cubic(xx), '--', label='Global Cubic Polynomial')
plt.legend()
plt.title("Cubic Spline vs Global Cubic Overshooting")
plt.grid(True)
plt.show()
```
